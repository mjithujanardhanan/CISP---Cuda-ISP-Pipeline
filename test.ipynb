```python
import numpy as np

# Your extracted array (Ordered: Grayscale -> CMY/RGB -> Misc -> Natural)
k = values

# 1. Reorder 'k' to match the standard Macbeth layout (Row 1 -> Row 4)
source_reordered = np.concatenate([k[18:24], k[12:18], k[6:12], k[0:6]])

# 2. Normalize linear sensor data to [0.0, 1.0]
# Because this is fresh from debayering, we skip the inverse sRGB gamma formula.
source_linear = source_reordered / 255.0

# 3. Standard Reference Linear sRGB values (D65)
reference_linear = np.array([
    [0.153, 0.086, 0.061], [0.443, 0.287, 0.226], [0.126, 0.198, 0.323], [0.095, 0.147, 0.061], 
    [0.224, 0.220, 0.419], [0.137, 0.498, 0.407], [0.640, 0.198, 0.040], [0.081, 0.106, 0.372], 
    [0.407, 0.095, 0.122], [0.107, 0.048, 0.153], [0.372, 0.488, 0.076], [0.686, 0.347, 0.040],
    [0.034, 0.054, 0.278], [0.126, 0.291, 0.081], [0.312, 0.035, 0.044], [0.781, 0.575, 0.036], 
    [0.380, 0.095, 0.291], [0.054, 0.247, 0.372], [0.905, 0.905, 0.865], [0.575, 0.581, 0.575], 
    [0.353, 0.353, 0.353], [0.191, 0.195, 0.191], [0.091, 0.091, 0.091], [0.031, 0.031, 0.031]
])

# 4. Compute the 3x3 CCM using Moore-Penrose pseudo-inverse (Least Squares)
ccm, residuals, rank, s = np.linalg.lstsq(source_linear, reference_linear, rcond=None)

# 5. Column-normalize to preserve White Balance
# This ensures that neutral grays (R=G=B) stay perfectly neutral after matrix multiplication.
ccm_normalized = ccm / np.sum(ccm, axis=0)

print("Your Calculated 3x3 Color Correction Matrix (CCM):\n")
print(np.array2string(ccm_normalized, formatter={'float_kind': lambda x: f"{x: .4f}"}))
```

```python
# Assuming 'Input' is your original loaded image
# image = Input.copy() 
# image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

# Resize for display
x_dim = int(image.shape[1] )
y_dim = int(image.shape[0] )
image = cv2.resize(image, (x_dim, y_dim))

# We need a clean copy to draw temporary rectangles over without ruining the base image
image_display = image.copy()

values = []
drawing = False
start_pt = (-1, -1)

def mouse_callback(event, x, y, flags, param):
    global drawing, start_pt, image_display, image
    
    # 1. Left click pressed: Start recording the area
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        start_pt = (x, y)
        
    # 2. Mouse moving: Draw a temporary rectangle to guide the user
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            # Refresh from the base image so we don't draw an infinite trail of boxes
            image_display = image.copy()
            cv2.rectangle(image_display, start_pt, (x, y), (0, 255, 0), 1)
            cv2.imshow("Image", image_display)
            
    # 3. Left click released: Finalize the area and calculate the average
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        end_pt = (x, y)
        
        # Ensure coordinates are ordered correctly (top-left to bottom-right)
        x1, x2 = min(start_pt, end_pt), max(start_pt, end_pt)
        y1, y2 = min(start_pt, end_pt), max(start_pt, end_pt)
        
        # Check if it was just a single click (no area dragged)
        if x1 == x2 or y1 == y2:
            # Fallback to single pixel logic
            b, g, r = image[y, x]
            print(f"Single Pixel Position: ({x}, {y})")
        else:
            # Slice the numpy array to extract the Region of Interest (ROI)
            roi = image[y1:y2, x1:x2]
            
            # Calculate the average (cv2.mean returns a 4-tuple, we slice the first 3 for B,G,R)
            avg_bgr = cv2.mean(roi)[:3]
            b, g, r = [int(v) for v in avg_bgr]
            
            print(f"Selected Area: Start({x1}, {y1}) to End({x2}, {y2})")

        print(f"Average RGB: ({r}, {g}, {b})")
        print("-" * 30)

        values.append((r, g, b))
        
        # Draw a permanent red rectangle on the base image to mark the spot
        cv2.rectangle(image, (start_pt, start_pt), (end_pt, end_pt), (0, 0, 255), 2)
        image_display = image.copy()
        cv2.imshow("Image", image_display)


cv2.namedWindow("Image")
cv2.setMouseCallback("Image", mouse_callback)

cv2.imshow("Image", image_display)

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # ESC to exit
        break

cv2.destroyAllWindows()
```
